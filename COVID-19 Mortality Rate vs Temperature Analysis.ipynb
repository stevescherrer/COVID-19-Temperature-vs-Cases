## Project Dependencies

```python
import os
import time
import datetime
import pandas as pd
import json
import urllib.request
import statsmodels.api as sm
import statsmodels.formula.api as smf
```

```python
## DarkSkyAPI Secret Key
darkSkyAPIKey = open('DarkSkyAPIKey.rtf').read().split(' ')[3].split('}')[0]
```

```python
## Defining Utility Functions
def glmR2(modelObj):
    """
    Calculate pseudo R2 of GLM: 1-(deviance/null.deviance)
    Input: Fitted GLM model from stastmodels.formula.api
    Output: Pseudo R2 value (0-1)
    """
    return (1 - (modelObj.deviance / modelObj.null_deviance))

def read_covid_file(filename):
    """
    Function for loading each of the 3 datasets
    """
    fData = pd.read_csv(os.path.join(dataDir, 'novel-corona-virus-2019-dataset/' + filename))
    ### Emphasis: italy, Spain, germany, France, and uk if possible
    fData = fData.loc[fData['Country/Region'].isin(['Italy', 'Spain', 'Germany', 'France', 'United Kingdom']), ]
    fData = fData.loc[fData['Province/State'].isna() == True, ] # 
    fData = fData.reset_index(drop = True) # Reindex remaining rows
    return fData
```

### Workspace Setup

```python
dataDir = os.path.join(os.getcwd(), 'Data')
```

```python
## Reading in Datafile
confirmedData = read_covid_file('time_series_covid_19_confirmed.csv')
recoveredData = read_covid_file('time_series_covid_19_recovered.csv')
deathData = read_covid_file('time_series_covid_19_deaths.csv')
```

### Getting Temperature Data for Affected Regions

```python
## Parsing out Dates
format_str = '%m/%d/%y'
dates = [datetime.datetime.strptime(X, format_str) for X in confirmedData.columns[4:]]
unixTimeStamps = [int(X.timestamp()) for X in dates]
```

```python
### Check to see if datafile exists, if it doesn't build a new one, otherwise load the previous file
if os.path.exists(os.path.join(dataDir, 'tempData.csv')):
    tempData = pd.read_csv(os.path.join(dataDir, 'tempData.csv'))
    
else:
    ### Collecting weather data using Dark Sky's API
    tempData = pd.DataFrame()

    for i in range(len(confirmedData['Country/Region'].unique())):
        lat = confirmedData.Lat[i]
        lon = confirmedData.Long[i]
        for j in range(len(dates)):
            ## Querey Darksky API 
            url = 'https://api.darksky.net/forecast/' + darkSkyAPIKey + '/' + str(lat) + ',' + str(lon) + ',' + str(unixTimeStamps[j])
            response = urllib.request.urlopen(url)

            ## Load requested data into a JSON file, then convert to pandas dataframe
            apiData = pd.io.json.json_normalize(json.loads(response.read()))

            ## Extract Daily Summary Data
            dailySummary = pd.DataFrame(apiData['daily.data'][0])

            ## Convert Time - Note: Time returned by API is always in local time for the location of the request
            dailySummary['time'] = dailySummary.time.apply(lambda X: time.gmtime(X))
            dailySummary = dailySummary.filter(['time', 'temperatureLow', 'temperatureHigh', 'apparentTemperatureLow', 'apparentTemperatureHigh'])

            ## Appending new columns
            dailySummary['lat'] = lat
            dailySummary['long'] = lon
            dailySummary['date'] = dates[j]
            dailySummary['country'] = confirmedData['Country/Region'].unique()[i]


            ### Getting Case Data
            dailySummary['confirmed'] = confirmedData.iloc[i, 4+j]
            dailySummary['recovered'] = recoveredData.iloc[i, 4+j]
            dailySummary['dead'] = deathData.iloc[i, 4+j]

            ## Appending Data
            tempData = pd.concat([tempData, dailySummary])

    ## Reorder dataframe columns        
    tempData = tempData[['country', 'date', 'lat', 'long', 'time', 'temperatureLow', 'temperatureHigh', 'apparentTemperatureLow', 'apparentTemperatureHigh', 'confirmed', 'recovered', 'dead']]

    # Reindex data for ARIMA
    tempData.index = tempData.date
    
    ### Feature Engineering
    ## Temperature Differentials
    tempData['diff'] = tempData['temperatureHigh'] - tempData['temperatureLow']
    tempData['apparentDiff'] = tempData['apparentTemperatureHigh'] - tempData['apparentTemperatureLow']

    ## Survival outcomes
    tempData['outcomeRatio'] = tempData['dead'] / tempData['recovered']
    tempData['survivalRate'] = tempData['recovered'] / (tempData['recovered'] + tempData['dead'])

    ## Save dataframe to data directory
    tempData.to_csv(os.path.join(dataDir, 'tempData.csv'))

    #### NOTE: IF this chunk keeps error-ing out, make sure line 10 is not commented out.
```
